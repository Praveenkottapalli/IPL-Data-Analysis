```python
# top 10 scorers in the tournament

top_run_scorers = deliveries_data.groupby("batsman")["batsman_runs"].sum().reset_index()
top_run_scorers = top_run_scorers.sort_values(by="batsman_runs", ascending=False)
top_run_scorers.head(10)
```

```python
# Count valid dismissals (excluding "run out" as it is not credited to a bowler)
valid_dismissals = ["caught", "bowled", "lbw", "stumped", "caught and bowled", "hit wicket"]

# Filter out valid dismissals and count wickets per bowler
top_wicket_takers = deliveries_data[deliveries_data["dismissal_kind"].isin(valid_dismissals)].groupby("bowler").size().reset_index(name="wickets")

# Sort in descending order to get top wicket takers
top_wicket_takers = top_wicket_takers.sort_values(by="wickets", ascending=False)

# Display top 10 wicket takers
print(top_wicket_takers.head(10))
```

```python
# Count total runs conceded per bowler
bowler_runs = deliveries_data.groupby("bowler")["total_runs"].sum().reset_index()

# Count total balls bowled per bowler
bowler_balls = deliveries_data.groupby("bowler")["ball"].count().reset_index()

# Merge the two dataframes
bowler_stats = pd.merge(bowler_runs, bowler_balls, on="bowler")

# Convert balls into overs (1 over = 6 balls)
bowler_stats["overs"] = bowler_stats["ball"] // 6  # Ignore incomplete overs

# Calculate economy rate
bowler_stats["economy_rate"] = bowler_stats["total_runs"] / bowler_stats["overs"]

# Sort by best economy (ascending order)
bowler_stats = bowler_stats.sort_values(by="economy_rate")
print(bowler_stats.head(10))  # Top 10 economical bowlers
```
